### Шаг 2. Выполните предобработку данных

#### Замена пропущенных значений

```python
print(real_estate_data.isna().sum())

'''
Значения пропущены в: ceiling_height, floors_total, living_area, is_apartment, kitchen_area, balcony, locality_name, airports_nearest, cityCenters_nearest, parks_around3000,
parks_nearest, ponds_around3000, ponds_nearest, days_exposition
'''
# Присвоим каждой квартире ценовую категорию, чтобы заменить недостающие значения средними значениями квартир той же ценовой категории
def set_price_category (last_price):
    if last_price <= 3400000:
        return 'low'
    elif 3400000 < last_price < 6800000:
        return 'medium'
    elif last_price >= 6800000:
        return 'high'
real_estate_data['price_category'] = real_estate_data['last_price'].apply(set_price_category)

# Заполним высоту потолков средним значением в той же ценовой категории
mean_ceiling_height_low = real_estate_data.loc[real_estate_data['price_category'] == 'low', 'ceiling_height'].mean()
mean_ceiling_height_medium = real_estate_data.loc[real_estate_data['price_category'] == 'medium', 'ceiling_height'].mean()
mean_ceiling_height_high = real_estate_data.loc[real_estate_data['price_category'] == 'high', 'ceiling_height'].mean()


def fill_ceiling_height (row):
    if pd.isna(row['ceiling_height']):
        if row['price_category'] == 'low':
            return mean_ceiling_height_low.round(2)
        elif row['price_category'] == 'medium':
            return mean_ceiling_height_medium.round(2)
        elif row['price_category'] == 'high':
            return mean_ceiling_height_high.round(2)
    else:
        return row['ceiling_height']
    
real_estate_data['ceiling_height'] = real_estate_data.apply(fill_ceiling_height, axis=1)

# Удалим строки, где floors_total == null
real_estate_data = real_estate_data.dropna(subset=['floors_total'])

# Заполним living_area медианным значением для каждой ценовой категории
median_living_area_low = real_estate_data.loc[real_estate_data['price_category'] == 'low', 'living_area'].median()
median_living_area_medium = real_estate_data.loc[real_estate_data['price_category'] == 'medium', 'living_area'].median()
median_living_area_high = real_estate_data.loc[real_estate_data['price_category'] == 'high', 'living_area'].median()

def fill_living_area(row):
    if pd.isna(row['living_area']):
        if row['price_category'] == 'low':
            return median_living_area_low
        elif row['price_category'] == 'medium':
            return median_living_area_medium
        elif row['price_category'] == 'high':
            return median_living_area_high
    else:
        return row['living_area']

real_estate_data['living_area'] = real_estate_data.apply(fill_living_area, axis=1)
# Исключим is_apartment из анализа из-за большого количества пропусков
real_estate_data = real_estate_data.drop('is_apartment', axis=1)
# Заполним kitchen_area медианным значением для каждой ценовой категории
median_kitchen_area_low = real_estate_data.loc[real_estate_data['price_category'] == 'low', 'kitchen_area'].median()
median_kitchen_area_medium = real_estate_data.loc[real_estate_data['price_category'] == 'medium', 'kitchen_area'].median()
median_kitchen_area_high = real_estate_data.loc[real_estate_data['price_category'] == 'high', 'kitchen_area'].median()

def fill_kitchen_area(row):
    if pd.isna(row['kitchen_area']):
        if row['price_category'] == 'low':
            return median_kitchen_area_low
        elif row['price_category'] == 'medium':
            return median_kitchen_area_medium
        elif row['price_category'] == 'high':
            return median_kitchen_area_high
    else:
        return row['kitchen_area']

real_estate_data['kitchen_area'] = real_estate_data.apply(fill_kitchen_area, axis=1)
# Заполним пустые значения balcony нулями
real_estate_data['balcony'] = real_estate_data['balcony'].fillna(0)
# Удалим строки, где locality_name неизвестно, поскольку таких строк очень мало
real_estate_data = real_estate_data.dropna(subset=['locality_name'])

# Для airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition оставим пропущенные значения, так как их нельзя заменить

real_estate_data
```

В датасете есть значения, которые заполняются пользователем, а есть значения, которые считаются автоматически.
Значения заполняемые пользователем могут иметь пропущенные значения из-за того, что эти поля были не обязательными для заполнения и пользователь просто пропустил их. Например, пользователь не знает точную площадь гостиной и кухни, поэтому не стал заполнять их.
Значения, которые заполняются автоматически, могут быть пропущены из-за багов в системе или же из-за недостатка информации от пользователя, чтобы рассчитать данные. Например, пользователь не указал название населённого пункта, поэтому расстояние до центра города и расстояние до аэропорта не посчитались.

#### Смена типов данных для столбцов

```python
def replace_to_int(columns):
    for column in columns:
        real_estate_data[column] = np.floor(pd.to_numeric(real_estate_data[column], errors='coerce')).astype('Int64')
        # real_estate_data[column] = real_estate_data[column].fillna(0)
        # real_estate_data[column] = real_estate_data[column].astype(int)

# Заменим floors_total, balcony, airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition
columns = ['floors_total', 'balcony', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'days_exposition']
replace_to_int(columns)
real_estate_data[columns].dtypes
```

Мы заменили тип данных с object (строки) на int, чтобы можно было проводить расчёты по этим столбцам. Замена значений с типа float на int нужна для удобства расчётов, так как int округляет значения

#### Обработка дубликатов

```python
# Удаление явных дубликатов
real_estate_data = real_estate_data.drop_duplicates()

print(f"Количество уникальных населённых пунктов до фильтрации: {real_estate_data['locality_name'].nunique()}")
# Удаление неявных дубликатов
real_estate_data['locality_name'] = real_estate_data['locality_name'].str.replace(
    r'поселок городского типа|поселок|посёлок', 'посёлок', regex=True)

print(f"Количество уникальных населённых пунктов после фильтрации: {real_estate_data['locality_name'].nunique()}")
```

### Шаг 3. Добавьте в таблицу новые столбцы

```python
# Добавляем цену за 1 квадратный метр
real_estate_data['price_sqm'] = round(real_estate_data['last_price'] / real_estate_data['total_area'], 2)

# Добавляем день недели
real_estate_data['first_day_exposition'] = pd.to_datetime(
    real_estate_data['first_day_exposition'],
    format='%Y-%m-%dT%H:%M:%S'
)
real_estate_data['published_weekday'] = real_estate_data['first_day_exposition'].dt.weekday

# Добавляем месяц публикации
real_estate_data['published_month'] = real_estate_data['first_day_exposition'].dt.month

# Добавляем год публикации
real_estate_data['published_year'] = real_estate_data['first_day_exposition'].dt.year

# Добавляем тип этажа квартиры
def set_floor_type (row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
real_estate_data['floor_type'] = real_estate_data.apply(set_floor_type, axis=1)

# Добавляем расстояние до центра города в километрах
real_estate_data['cityCenters_nearest'] = pd.to_numeric(real_estate_data['cityCenters_nearest'], errors='coerce')
real_estate_data['cityCenters_nearest_km'] = (real_estate_data['cityCenters_nearest'] / 1000).astype(float).round()

real_estate_data['cityCenters_nearest'] = pd.to_numeric(real_estate_data['cityCenters_nearest'], errors='coerce')
real_estate_data['cityCenters_nearest_km'] = (real_estate_data['cityCenters_nearest'] / 1000).astype(float).round()

real_estate_data.head()
```

Средняя цена одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений

```python
top_10_by_amount = real_estate_data.groupby('locality_name').size().sort_values(ascending=False).head(10)

top_10_by_amount_data = pd.pivot_table(real_estate_data, index='locality_name', values='price_sqm', aggfunc=['count', 'mean'])

top_10_sqm = top_10_by_amount_data.loc[top_10_by_amount.index]

top_10_sqm['mean'].plot(kind='bar', title='Средняя цена одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений',
                       xlabel='Населённый пункт', ylabel='Среднеяя цена за квадратный метр')

most_expensive = top_10_sqm.iloc[0]
least_expensive = top_10_sqm.iloc[-1]

print(f"Самая высокая цена за квадратный метр: {int(most_expensive['mean']['price_sqm'])} в населённом пункте {most_expensive.name}")
print(f"Самая низкая цена за квадратный метр: {int(least_expensive['mean']['price_sqm'])} в населённом пункте {least_expensive.name}")
```
